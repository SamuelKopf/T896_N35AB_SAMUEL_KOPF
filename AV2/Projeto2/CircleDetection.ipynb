```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Carrega a imagem
img = cv2.imread('circulos_1.png')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
# Suavização com filtro Gaussiano
blur = cv2.GaussianBlur(gray, (9, 9), 2)
```

```python
# Detecção de círculos com Hough
circles = cv2.HoughCircles(
    blur,
    cv2.HOUGH_GRADIENT,
    dp=1.2,
    minDist=80,
    param1=100,
    param2=30,
    minRadius=55,
    maxRadius=70
)
```

```python
# Desenho dos círculos encontrados
resultado = img.copy()
if circles is not None:
    circles = np.uint16(np.around(circles))
    for (x, y, r) in circles[0, :]:
        cv2.circle(resultado, (x, y), r, (0, 255, 0), 3)
```

```python
# Contagem de círculos detectados
num_circles = len(circles[0]) if circles is not None else 0
```

```python
# Exibição do resultado
plt.figure(figsize=(8, 8))
plt.imshow(cv2.cvtColor(resultado, cv2.COLOR_BGR2RGB))
plt.title(f"Circulos detectados: {num_circles}")
plt.axis("off")
plt.show()
```
